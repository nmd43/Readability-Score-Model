## Conducting PCA on Readability Scores

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('scores.csv')
features = data.columns.to_numpy()
#features_labels = np.append(features,'label')
#data.columns = features_labels
```

```python
x = data.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_scores = pd.DataFrame(x,columns=feat_cols)
#print(normalised_scores.tail())
```

```python
from sklearn.decomposition import PCA
pca_scores_2 = PCA(n_components=2)
principalComponents_scores_2 = pca_scores_2.fit_transform(x)

principal_scores_Df2 = pd.DataFrame(data = principalComponents_scores_2, columns = ['principal component 1', 'principal component 2'])
#print(principal_scores_Df2.tail())
print('Explained variation per principal component: {}'.format(pca_scores_2.explained_variance_ratio_))
```

```python
import plotly.express as px
total_var_2 = pca_scores_2.explained_variance_ratio_.sum() * 100
fig = px.scatter(principalComponents_scores_2, x=0, y=1, title=f'Total Explained Variance: {total_var_2:.2f}%')
fig.show()
```

```python
pca_scores_3 = PCA(n_components=3)
principalComponents_scores_3 = pca_scores_3.fit_transform(x)

principal_scores_Df3 = pd.DataFrame(data = principalComponents_scores_3, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
#print(principal_scores_Df3.tail())
print('Explained variation per principal component: {}'.format(pca_scores_3.explained_variance_ratio_))
total_var_3 = pca_scores_3.explained_variance_ratio_.sum() * 100

fig2 = px.scatter_3d(
    principalComponents_scores_3, x=0, y=1, z=2,
    title=f'Total Explained Variance: {total_var_3:.2f}%',
    labels={'0': 'PC 1', '1': 'PC 2', '2': 'PC 3'})
fig2.update_traces(marker={'size': 3})
fig2.show()
```
