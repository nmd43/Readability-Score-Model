```python
from ngram import * 
import os 
import matplotlib.pyplot as plt 

files = sorted(os.listdir("/home/mbahng/Desktop/dataplus-2023-techdoc/data"))
```

```python
PQIs = []

for file in files: 
    model = NgramModel(4, [f"/home/mbahng/Desktop/dataplus-2023-techdoc/data/{file}"])
    PQIs.append(model.PQIndex(3, 2))
```

```python
plt.hist(PQIs, bins=100, edgecolor="black") 
plt.xlabel("PQ Index (p=3, q=2)")
plt.ylabel("Number of Documents")
plt.show()
```
